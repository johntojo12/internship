```python
 for date in trDate:
    for x in location_lst:
        for y in location_lst:
            if x!=y: 
                baseDataUrl = "https://www.makemytrip.com/flight/search?itinerary="+ x +"-"+ y +"-"+ date +"&tripType=O&paxType=A-1_C-0_I-0&intl=false&cabinClass=E&ccde=IN&lang=eng"
                driver.get(baseDataUrl)

                time.sleep(4)
                print ('*'*30)
                for j in range(1, 100):
                    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

                time.sleep(4)


                FlightName = driver.find_elements(By.XPATH, '//p[@class="boldFont blackText airlineName"]')
                FlightName = [n.text for n in FlightName]
                name.append ([flight_name for flight_name in FlightName])
                

                
                
                duration_min=[]
                FlightDuration = driver.find_elements(By.XPATH, '//div[@class="stop-info flexOne"]/p')
                FlightDuration = [d.text for d in FlightDuration]
                for t in FlightDuration:
                    if 'h' in t:
                        # Split the time string by 'h'
                        hours, minutes = t.split('h')

                        # Convert hours to minutes and add to the minutes value
                        total_minutes = int(hours) * 60
                    else:
                        # If the string does not contain 'h', assume it is in the format 'mm'
                        minutes = t
                        total_minutes = 0
                    
                    # If the second element contains 'm', split it by 'm' and add the minutes value
                    if 'm' in minutes:
                        total_minutes += int(minutes.split('m')[0])
                    duration_min.append(total_minutes)
                duration_minute.append(duration_min)



                FlightPrice = driver.find_elements(By.XPATH, '//p[@class="blackText fontSize18 blackFont white-space-no-wrap"]')
                FlightPrice = [p.text for p in FlightPrice]
                for p in FlightPrice:
                    price.append(p)
                 
                
                FlightStop = driver.find_elements(By.XPATH, '//p[@class="flightsLayoverInfo"]')
                FlightStop = [s.text for s in FlightStop]
                for stop in FlightStop:
                    if stop:  # Check if the string is not empty
                        if stop != 'Non stop':
                            # Convert the first character to an integer
                            stops.append(int(stop[0]))
                        else:
                            stops.append(0)  # Set the number of stops to 0 if the flight is non-stop
                    else:
                        stops.append(np.nan)  # Set the number of stops to 0 if the string is empty

                


                FligtArDpTime = driver.find_elements(By.XPATH, '//p[@class="appendBottom2 flightTimeInfo"]')
                FligtArDpTime =[A.text for A in FligtArDpTime]
                clean_times = [t.replace('\n+ 1 DAY', '') for t in FligtArDpTime]
                departure_time.append( [t for i, t in enumerate(clean_times) if i % 2 == 0])
                arrival_time.append ([t for i, t in enumerate(clean_times) if i % 2 == 1])
              


                
                FlightDpArLoc = driver.find_elements(By.XPATH, '//p[@class="blackText"]')
                FlightDpArLoc = [n.text for n in FlightDpArLoc]
                departure_loc .append([t for i, t in enumerate(FlightDpArLoc) if i % 2 == 0])
                arrival_loc .append([t for i, t in enumerate(FlightDpArLoc) if i % 2 == 1])
     
                
                current_date = date
                for i in range(len(FlightName)):
                    if date != current_date:
                        current_date = date
                    Date.append(current_date)
                    


                
                print(x,' to ',y, ' date ',date)
                print(len(FlightPrice))
               
```

```python
def remove_rupee_and_comma(values):
    new_values = []
    for value in values:
        value = value.replace("₹", "")
        value = value.replace(",", "")
        new_values.append(value)
    return new_values
```
