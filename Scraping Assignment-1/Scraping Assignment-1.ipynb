# q2 Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
url='https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating'
r=requests.get(url)
html_content=r.text
soup=BeautifulSoup(html_content,'html.parser')
```

```python
name=[]
year=[]
rating=[]

movie=soup.findAll('div', class_='lister-item mode-advanced')
for m in movie:
    name.append(m.h3.a.text)
    rate = m.find('div', class_ = 'inline-block ratings-imdb-rating').text.replace('\n', '')
    rating.append(rate)
    year_of_release = m.h3.find('span', class_ = 'lister-item-year text-muted unbold').text.replace('(', '').replace(')', '')
    year.append(year_of_release)

```

```python
movie_DF = pd.DataFrame({'Name of movie': name,'Movie Rating': rating ,'Year of relase': year})
movie_DF
```

# q4 Write s python program to display list of respected former presidents of India(i.e. Name , Term of office) from https://presidentofindia.nic.in/former-presidents.htm

```python
url='https://presidentofindia.nic.in/former-presidents.htm'
r=requests.get(url)
html_content=r.text
soup=BeautifulSoup(html_content,'html.parser')
```

```python
name=[]
year=[]
presidents=soup.findAll('div', class_='presidentListing')
for t in presidents:
    b=t.h3.text
    c= b.split("(")
    name.append(c[0])
    year_a = t.find('p').text
    x = year_a.split(": ")
    year.append(x[1])
```

```python
presidents = pd.DataFrame({'Name of president': name, 'Term of office': year})
presidents
```

# q5 Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:

a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
url='https://www.icc-cricket.com/rankings/mens/team-rankings/odi'
r=requests.get(url)
html_content=r.text
soup=BeautifulSoup(html_content,'html.parser')
```

```python
matches=[]
points=[]
rating=[]
team=[]

cricket=soup.findAll('tr', class_='rankings-block__banner')
for c in cricket:
    b=c.find('span',class_='u-hide-phablet').text
    team.append(b)
    d=c.find('td',class_='rankings-block__banner--matches').text
    matches.append(d)
    e=c.find('td',class_='rankings-block__banner--points').text
    points.append(e)
    
    f=c.find('td',class_='rankings-block__banner--rating u-text-right').text
    for z in f.split():
        if z.isdigit():
            rating.append(int(z))
        


cricket=soup.findAll('tr', class_='table-body')
for c in cricket:
    b=c.find('span',class_='u-hide-phablet').text
    team.append(b)
    d=c.find('td',class_='table-body__cell u-center-text').text
    matches.append(d)
    f=c.find('td',class_='table-body__cell u-text-right rating').text
    rating.append(f) 
    
h4=[] #used because there are two tags having same name to extract points
for i in soup.find_all('td',class_='table-body__cell u-center-text'):
    h4.append(i.text)
    
for i in range(len(h4)):
    if(i%2!=0):
        points.append(h4[i])
```

```python
a= pd.DataFrame({ 'Team': team,'Matches':matches,'Points':points,'Ratings':rating})
a.drop(a.index[10:], inplace=True)
a 
```

# q7 Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world :i) Headline ii) Time iii) News Link

```python
url='https://www.cnbc.com/world/?region=world'
r=requests.get(url)
html_content=r.text
soup=BeautifulSoup(html_content,'html.parser')
```

```python
head=[]
time=[]
url=[]
news=soup.findAll('li', class_='LatestNews-item')
for n in news:
    b=n.find('a',class_='LatestNews-headline').text
    head.append(b)
    b=n.find('time',class_='LatestNews-timestamp').text
    time.append(b)
    

b=soup.findAll('a',class_='LatestNews-headline')
for k in b:
    if 'href' in k.attrs:
         url.append(str(k['href']))  
```

```python
a= pd.DataFrame({ 'Headline':head,'Time':time,'NewsLink':url})
a
```

# q9 Write a python program to scrape mentioned details from dineout.co.in : i) Restaurant name ii) Cuisine iii) Location iv) Ratings v) Image URL

```python
url='https://www.dineout.co.in/delhi-restaurants/buffet-special'
r=requests.get(url)
html_content=r.text
soup=BeautifulSoup(html_content,'html.parser')
name=[]
cuisine=[]
loc=[]
rating=[]
iurl=[]
thumbnail_elements = soup.find_all("img", class_ = "no-img")
for element in thumbnail_elements:
    print(str(element['data-src']))
    iurl.append(str(element['data-src']))
```

```python
rest=soup.findAll('div', class_='restnt-card restaurant')
for r in rest:
    b=r.find('div',class_='restnt-rating rating-4').text
    rating.append(b)
    b=r.find('a',class_='restnt-name ellipsis').text
    name.append(b)
    b=r.find('div',class_='restnt-loc ellipsis').text
    loc.append(b)
    b=r.find('span',class_="double-line-ellipsis").text
    b=b.split('|')
    cuisine.append(b[1])
    
```

```python
a= pd.DataFrame({ 'Restaurant name':name,'Cuisine':cuisine,'Location':loc,'Ratings':rating,'Image URL':iurl})
a 
```

# q10 Write a python program to scrape the details of top publications from Google Scholar from https://scholar.google.com/citations?view_op=top_venues&hl=en i) Rank ii) Publication iii) h5-index iv) h5-median

```python
url='https://scholar.google.com/citations?view_op=top_venues&hl=en'
r=requests.get(url)
html_content=r.text
soup=BeautifulSoup(html_content,'html.parser')
```

```python
rank=[]
pub=[]
h5=[]
h4=[] #while running the loop h4 returns value of h5-index and h5-median, h5-median values are stored in even places later extracted and stored in h
h=[]

for i in soup.find_all('td',class_='gsc_mvt_p'):
    rank.append(i.text)
for i in soup.find_all('td',class_='gsc_mvt_t'):
    pub.append(i.text)  
for i in soup.find_all('td',class_='gsc_mvt_n'):
    h4.append(i.text)      
for i in soup.find_all('span',class_='gs_ibl gsc_mp_anchor'):
    h.append(i.text)
for i in range(len(h4)):
    if(i%2==0):
        h5.append(h4[i])
```

```python
F = pd.DataFrame({'rank': rank, 'Publication': pub,  'h5-index': h5,'h5-median':h})
F
```
